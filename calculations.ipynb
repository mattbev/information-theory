```python
import numpy as np
import matplotlib.pyplot as plt


payouts = [35, 17, 11, 8, 6, 5, 2, 2, 1, 1, 1, 1, 1, 1]
subfair_odds = np.array([
    .026,
    .053,
    .079,
    .105,
    .132,
    .158,
    .316,
    .316,
    .474,
    .474,
    .474,
    .474,
    .474,
    .474
])

fair_odds = []
for i in payouts:
    fair_odds.append(1/(i+1)) 
fair_odds = np.array(fair_odds)
# print(fair_odds)

delta = fair_odds - subfair_odds
print(delta)
betnames = [
    "Single",
    "Two",
    "Three",
    "Four",
    "Five",
    "Six",
    "Column",
    "Dozen",
    "Even",
    "Odd",
    "Red",
    "Black",
    "Low",
    "High"
]
barlist = plt.bar([i for i in range(14)], delta, label="Delta Values")
plt.xticks([i for i in range(14)], betnames)
plt.title("Difference between Implied and Actual Odds")
plt.xlabel("bet")
plt.ylabel("difference")
plt.rcParams["figure.figsize"] = (10,5)
barlist[5].set_color('r')
plt.savefig("deltas")
plt.show()
```

```python
from random import randint

class Bet:
    def __init__(self, b):
        self.one = b[0]
        self.two = b[1]
        self.three = b[2]
        self.four = b[3]
        self.five = b[4]
        self.six = b[5]
        self.column = b[6]
        self.dozen = b[7]
        self.even = b[8]
        self.odd = b[9]
        self.red = b[10]
        self.black = b[11]
        self.low = b[12]
        self.high = b[13]
        
class Player:
    def __init__(self, cash):
        self.cash = cash
        self.cash_history = [cash]
        
    def make_bet(self, bet):
        doz_ind = randint(0,2)
        
        self.bets = {
            "one" : (np.random.choice(Roulette.pockets, size=1, replace=False), bet.one), 
            "two" : (np.random.choice(Roulette.pockets, size=2, replace=False), bet.two),
            "three" : (np.random.choice(Roulette.pockets, size=3, replace=False), bet.three),
            "four" : (np.random.choice(Roulette.pockets, size=4, replace=False), bet.four),
            "five" : (np.random.choice(Roulette.pockets, size=5, replace=False), bet.five),
            "six" : (np.random.choice(Roulette.pockets, size=6, replace=False), bet.six),
            "column" : (Roulette.pockets[randint(2,4)::3], bet.column),        
            "dozen" : (Roulette.pockets[2+doz_ind*12: 2+(doz_ind+1)*12], bet.dozen),
            "even" : (Roulette.pockets[2:][np.where(Roulette.pockets[2:].astype(int) % 2 == 0)], bet.even),
            "odd" : (Roulette.pockets[2:][np.where(Roulette.pockets[2:].astype(int) % 2 != 0)], bet.odd),
            "red" : (np.array([1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36]).astype(str), bet.red),
            "black" : (np.array([2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33, 35]).astype(str), bet.black),
            "low" : (Roulette.pockets[2:20], bet.low),
            "high" : (Roulette.pockets[20:], bet.high)
        }
        
        return self.bets 
    
    def check_bets_and_update_cash(self, winning_pocket):
        difference = 0
        for bet in self.bets:
            selected_pockets, b = self.bets[bet]
            if winning_pocket in selected_pockets:
                difference += Roulette.payouts[bet]*b*self.cash
            else:
                difference -= b*self.cash
        self.cash += difference
        self.cash_history.append(self.cash)
    
class Roulette:
    pockets = np.array(["00"] + [str(i) for i in range(37)])
    payouts = {
        "one" : 35, 
        "two" : 17,
        "three" : 11,
        "four" : 8,
        "five" : 6,
        "six" : 5,
        "column" : 2,        
        "dozen" : 2,
        "even" : 1,
        "odd" : 1,
        "red" : 1,
        "black" : 1,
        "low" : 1,
        "high" : 1
    } 
    
    @staticmethod
    def spin(players, num_spins=1):
        for spin in range(num_spins):
            winning_pocket = np.random.choice(Roulette.pockets, size=1)
            for player in players:
                player.check_bets_and_update_cash(winning_pocket)
        
```

```python
import time

trials = 1000
kelly_histories = []
delta_histories = []
kt_histories = []
kt_opt_histories = []

start = time.time()

for trial in range(trials):
    cash = 1000
    num_spins = 20

    kelly_player = Player(cash)
    kelly_props = subfair_odds/np.sum(subfair_odds)
    kelly_bet = Bet(kelly_props)
    kelly_player.make_bet(kelly_bet)

    delta_player = Player(cash)
    delta_props = (1-(delta)/np.sum(delta))*kelly_props
    delta_props = delta_props/np.sum(delta_props)
    delta_bet = Bet(delta_props)
    delta_player.make_bet(delta_bet)
    
    kt_player = Player(cash)
    kt_props = kelly_props
    kt_props_min = min([kt_props[i]*list(Roulette.payouts.values())[i] for i in range(len(kt_props))])
    kt_props = np.array([kt_props[i] - kt_props_min/list(Roulette.payouts.values())[i] for i in range(len(kt_props))])
    kt_bet = Bet(kt_props)
    kt_player.make_bet(kt_bet)
    
    kt_opt_player = Player(cash)
    kt_opt_props = delta_props
    kt_opt_props_min = min([kt_opt_props[i]*list(Roulette.payouts.values())[i] for i in range(len(kt_opt_props))])
    kt_opt_props = np.array([kt_opt_props[i] - kt_opt_props_min/list(Roulette.payouts.values())[i] for i in range(len(kt_opt_props))])
    kt_opt_bet = Bet(kt_opt_props)
    kt_opt_player.make_bet(kt_opt_bet)
    
    players = [kelly_player, delta_player, kt_player, kt_opt_player]
    Roulette.spin(players=players, num_spins=num_spins)
    
    kelly_histories.append(kelly_player.cash_history)
    delta_histories.append(delta_player.cash_history)
    kt_histories.append(kt_player.cash_history)
    kt_opt_histories.append(kt_opt_player.cash_history)

kelly_histories = np.stack(kelly_histories)
delta_histories = np.stack(delta_histories)
kt_histories = np.stack(kt_histories)
kt_opt_histories = np.stack(kt_opt_histories)

kelly_avg = np.mean(kelly_histories)
delta_avg = np.mean(delta_histories)
kt_avg = np.mean(kt_histories)
kt_opt_avg = np.mean(kt_opt_histories)

kelly_slopes = np.apply_along_axis(np.gradient, 1, kelly_histories)
delta_slopes = np.apply_along_axis(np.gradient, 1, delta_histories)
kt_slopes = np.apply_along_axis(np.gradient, 1, kt_histories)
kt_opt_slopes = np.apply_along_axis(np.gradient, 1, kt_opt_histories)

kelly_slope_avg = np.mean(kelly_slopes)
delta_slope_avg = np.mean(delta_slopes)
kt_slope_avg = np.mean(kt_slopes)
kt_opt_slope_avg = np.mean(kt_opt_slopes)

end = time.time()

print("Average Proportion of Starting Cash in Pocket")
print("-"*50)
print("Kelly Gambling: {0}".format(kelly_avg/cash))
print("Delta Gambling: {0}".format(delta_avg/cash))
print("KT Gambling: {0}".format(kt_avg/cash))
print("KT Opt Gambling: {0}".format(kt_opt_avg/cash))


print("\n")

print("Average Rate of Decay")
print("-"*50)
print("Kelly Gambling: {0}".format(kelly_slope_avg/cash))
print("Delta Gambling: {0}".format(delta_slope_avg/cash))
print("KT Gambling: {0}".format(kt_slope_avg/cash))
print("KT Opt Gambling: {0}".format(kt_opt_slope_avg/cash))

print("\n")

print("Time Elapsed: {0} seconds".format(end-start))
```
